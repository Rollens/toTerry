```python
import csv
from matplotlib import pyplot as plt
import os
from collections import defaultdict
```

```python
data_list = os.listdir('data')
```

```python
dataset = {}
for d in data_list:
    route = os.path.join('data',d)
    with open(route,'r',encoding='big5') as f:
        rows = csv.reader(f)
        flag = False # find PosX or not
        temp=[]
        for row in rows:
            if row:
                if row[0] == 'PosX':
                    flag = True
                    try:
                        lop2_index = row.index('LOP2')
                    except:
                        print('LOP2 is not in list.')
                elif flag:
                    temp.append(float(row[lop2_index]))
    dataset[d[:-4]] = temp
x_label = [*(range(len(temp)))] 
```

```python
plt.figure(figsize=(12,5))
plt.title('LOP2 graph demo')
for key , value in dataset.items():
    plt.scatter(x_label,value,label=key)
plt.ylim(0.25,0.4)
plt.grid(linestyle='--')
plt.legend()
plt.show()
```
